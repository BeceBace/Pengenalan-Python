# Plot Data dengan Menggunakan matplotlib
Untuk memplotting data pada dapat menggunakan modul matplotlib python dengan menggunakan atribut **pyplot**. fungsi2 dan atribut yang digunakan adalah:
**plot, title,  xlabel, ylabel**, dan fungsi **show()** untuk menampilkan plotting <br>
Contoh statement :
```python 
    import matplotlib.pyplot as  plt
    dataset.plot(x= .., y=.., style='o')  
    plt.title('X vs Y')  
    plt.xlabel('X')  
    plt.ylabel('Y')  
    plt.show()
    ```

# Subsetting DataFrame di Pandas
Untuk subsetting baris atau kolom saja biasanya menggunakan '[]' yang diisi dengan jangkauan baris dan jangkauan kolom, contoh:<br>
data2[:11] diatas <br>
Untuk subsetting baris dan kolom, biasanya menggunakan atribut **loc** dan **iloc**, <br>
**loc** : digunakan untuk subseting dengan menggunakan nama index dan nama kolom <br>
**iloc** : digunakan untuk subsetting dengan menggunakan posisi index dan posis kolom (berupa integer)<br>
Contoh : <br>
```python
    data1.loc[[1,3,4],['Open', 'High']]
    data1.iloc[[1,3,4],[0,1]]
    ```

````python
import pandas as pd
data1 = pd.read_csv('C:/Users/Putra/Downloads/GOOG(1).csv')
````

````python
data1.sample(5)
````

````python
data1.drop(['Date'], axis = 1) #data1 tanpa kolom 'Date', axis = 1 menunjukan axis kolom (axis = 0 , menunjukan axis baris)
````

````python
data2 = data1
for i in data2.drop(['Date'], axis = 1).columns :
    data2[i] = data2[i] / data2[i].sum() # ubah nilai data ke bentuk normalisasi
````

````python
data2
````

````python
import matplotlib.pyplot as plt
data2[['Open','Close']].plot(x = 'Open', y='Close', style = 'o')
plt.title('Open Vs Close')
plt.xlabel('X')
plt.ylabel('Y')
%matplotlib inline
````
